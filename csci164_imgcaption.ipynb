# Captioner Installation

```python
captioner1 = pipeline(task="image-to-text",
                      model="Salesforce/blip-image-captioning-base")
```

```python
captioner2 = pipeline("image-to-text",
                      model="seckmaster/microsoft-git-base")
```

```python
captioner3 = pipeline("image-to-text",
                      model="cnmoro/tiny-image-captioning")
```

# File #1 - Video Game Screenshot
The first file, being a screenshot from my upcoming game: RAYBOUND: SUBNOVIAN LEGEND. It depicts a square and a deer person sitting, staring at a statue in the background. This is not a real life photo, so the captioners might have trouble identifying them.

```python
display(Image(filename=file_name1))
```

```python
captioner1(file_name1)
```

```python
captioner2(file_name1)
```

```python
captioner3(file_name1)
```

Captioner 1 seems to be the most accurate regarding file #1.
